Para identificar as variáveis explicativas, é ideal analisar os dados com gráficos, histograma ou boxsplot.

```python
x = dados.drop('churn', axis=1)
```

```python
x
```

```python
y = dados['churn']
```

```python
from sklearn.compose import make_column_transformer
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
colunas = x.columns
```

```python
one_hot = make_column_transformer(
    (OneHotEncoder(drop='if_binary'),
     ['pais','sexo_biologico','tem_cartao_credito','membro_ativo']
     ),
    remainder='passthrough',
    sparse_threshold=0
)
```

```python
colunas
```

```python
colunas = x.columns
```

```python
colunas
```

```python
x = one_hot.fit_transform(x)
```

```python
one_hot.get_feature_names_out(colunas)
```

```python
pd.DataFrame(x, columns=one_hot.get_feature_names_out(colunas))
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
y = label_encoder.fit_transform(y)
```

Fazer o treinamento de dados e depois testar para ver se ocorreu aprendizado.

```python
from sklearn.model_selection import train_test_split
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, stratify=y, random_state=5)
```

```python
from sklearn.dummy import DummyClassifier
```

```python
dummy = DummyClassifier()
```

```python
dummy.fit(x_treino, y_treino)
dummy.score(x_teste, y_teste)
```

Aplicando árvore de decisão

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
arvore = DecisionTreeClassifier(random_state=5)
```

```python
arvore.fit(x_treino, y_treino)
```

```python
arvore.predict(x_teste)
```

```python
arvore.score(x_teste, y_teste)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
dados
```

```python
arvore = DecisionTreeClassifier(max_depth=4, random_state=5)
arvore.fit(x_treino, y_treino)
```

```python
plt.figure(figsize=(15,6))
plot_tree(arvore, filled=True, class_names=['Nao','Sim'],fontsize=7);
```

```python
arvore.score(x_treino, y_treino)
```

```python
arvore.score(x_teste, y_teste)
```

Normalizar os dados para o KNN

```python
from sklearn.preprocessing import MinMaxScaler
normalizacao = MinMaxScaler()
x_treino_normalizado = normalizacao.fit_transform(x_treino)
x_teste_normalizado = normalizacao.fit_transform(x_teste)
```

```python
pd.DataFrame(x_treino_normalizado)
```

Utilizando o KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
knn.fit(x_treino_normalizado, y_treino)
```

```python
knn.score(x_teste_normalizado, y_teste)
```

```python
print(f'Acurácia Dummy: {dummy.score(x_teste, y_teste)}')
print(f'Acurácia Árvore: {arvore.score(x_teste, y_teste)}')
print(f'Acurácia KNN: {knn.score(x_teste_normalizado, y_teste)}')
```
